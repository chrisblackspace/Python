```python
#Libraries
import jinja2
from pycaret.regression import *
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#Load th data set
from google.colab import files
files.upload()
```

```python
#Read in the data
asset = pd.read_csv('XRP-USD.csv')
#Set teh date as the index
asset = asset.set_index(pd.DatetimeIndex(asset['Date'].values))
#show the data
asset
```

```python
#make a copy of the data set
df3 = asset.copy()
#store the data in a feature data set call 'x' and add in the target column
X = np.array(df3[['Date', 'Open', 'Close']])
#Store the close price as an array in a variable called 'y'
y = df3['Close'].values
#split data 85% training and 15% testing 
x_train, x_test, y_train , y_test, = train_test_split(X, y, test_size=0.15, random_state =0, shuffle= False)
```

```python
#Get the train data and transform it to a data frame
train_data = pd.DataFrame(x_train, columns= ['Date', 'Open', 'Close'])
#Show the first 7 rows of data
train_data.head(7)
```

```python
#Get the test data and transform it into a data frame
test_data = pd.DataFrame(x_test, columns= ['Date', 'Open', 'Close'])
#Show the first 7 rows of data
test_data.head(7)
```

```python
#Initialize the setup
s = setup(data=train_data, target='Close', session_id = 123, use_gpu = True)
```

```python
#Train on multiple models and compare them 
compare_models(sort= 'MSE')
```

```python
#Create the model
en = create_model('en')
```

```python
#Get the predictions
predict_model(en, data=test_data.drop(['Close'], axis=1))
```
